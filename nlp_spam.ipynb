**Data cleaning and preprocessing**

```
#remove punctuation
#remove stop words
def process_text(text):
  nopunc=[char for char in text if char not in string.punctuation]
  nopunc=''.join(nopunc)
  clean_words=[word for word in nopunc.split()if word.lower() not in stopwords.words('english')]  
  return clean_words 
```

```
df['Text'].head().apply(process_text)
```

```
from sklearn.feature_extraction.text import CountVectorizer
bow1=CountVectorizer().fit_transform(df['Text']).toarray()
print(bow1)
```

```
#train test split

from sklearn.model_selection import train_test_split
#y=df.iloc[:,1].values
#x=bow1
X_train, X_test, y_train, y_test = train_test_split(bow1,df['Label'], test_size = 0.20, random_state = 0)
```

```
#classifier
from sklearn.naive_bayes import MultinomialNB
mnclassifier = MultinomialNB().fit(X_train, y_train)
```

```
y_pred=mnclassifier.predict(X_test)
print(y_pred)
```

```
print(y_test)
```

```
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy_score(y_test,y_pred)
```

```
#logistic regression
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='lbfgs')
clf.fit(X_train, y_train)

predlog=clf.predict(X_test)
print(predlog)
```

```
accuracy_score(y_test,predlog)
```
